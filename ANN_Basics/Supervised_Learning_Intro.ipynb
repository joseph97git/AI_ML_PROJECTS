```python
import matplotlib.pyplot as plt
import numpy as np
```

## The Basics
Supervised learning consists of getting <span style='color:red'>inputs to match a certain output</span>. In order to achieve this, we give the artifical neural network various inputs and compare it to our expected output, updating the weights according to how well the output compares to what we want. 

## Activation Function
In order to "turn on" a node we must include some sort of activation function. The one we usually use is the <span style='color:red'>sigmoid function</span>. 
<br>
<center>
    \begin{equation*}
        f(x) = 1 / (1 + exp(-x))
    \end{equation*}
</center>

```python
# plotting the sigmoid function
x = np.arange(-8, 8, 0.1)
f = 1 / (1 + np.exp(-x))
plt.plot(x, f)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Sigmoid Function')
plt.show()
```

## Bias
There are different weights associate with each node that determine the strength of the relationship between the input and the output. By increaseing the strength or weight or bias of the node, we are saying that the <span style='color:red'>input more strongly implies the output</span>. The by increasing the bias, we change the slope. A very large bias will increase the slope of the sigmoid function so much so that it will be a vertial line from 0 to 1, binary. Thus, the input directly implies the output.

```python
# plot showing how the bias affects the slope
w1 = 0.5
w2 = 1.0
w3 = 2.0
l1 = 'w = 0.5'
l2 = 'w = 1.0'
l3 = 'w = 2.0'
for w,l in [(w1, l1), (w2, l2), (w3, l3)]:
    f = 1 / (1 + np.exp(-x*w))
    plt.plot(x, f, label=l)
plt.xlabel('x')
plt.ylabel('h_w(x)')
plt.legend(loc=2)
plt.title('Bias Changes Slope')
plt.show()
```

## Feed Forward
We can use matrix algebra to represent a simple neural network with a weight matrix **W**, a bias vector **b**, and an input vector **x** to determine the output after one feed forward pass.

```python
def f(x):
    # sigmoid function
    return 1 / (1 + np.exp(-x))

def matrix_feed_forward_calc(n_layers, x, w, b):
    # Output after one feed forward pass
    # given the number of layers, an initial
    # input vector, x, a weight matrix, w, and
    # a bias vector, b, for a simple
    # artificial neural network
    for l in range(n_layers - 1):
        if l == 0:
            node_in = x
        else:
            node_in = h
        z = w[l].dot(node_in) + b[l]
        h = f(z)
    return h
```

```python
# feed forward example
w1 = np.array([[0.2, 0.2, 0.2], [0.4, 0.4, 0.4], [0.6, 0.6, 0.6]])
w2 = np.zeros((1, 3))
w2[0,:] = np.array([0.5, 0.5, 0.5])
b1 = np.array([0.8, 0.8, 0.8])
b2 = np.array([0.2])

w = [w1, w2]
b = [b1, b2]
x = [1.5, 2.0, 3.0]

matrix_feed_forward_calc(3, x, w, b)
```
